## Read in the data

<p>Read in the datafile in the <code>data/</code> directory which gives the number of corgis born in several countries over time, this datafile is called <code>corgis_per_country.csv</code>.  Call the dataframe <code>df_corg</code>.<p><p>Update the 'Years' column so that it is a time-like variable with: </p><p><code>df_corg['Years'] = pd.to_datetime(df_corg['Years'].astype('int'), format='%Y')<code>.</p>

```python
#grade (write your code in this cell and DO NOT DELETE THIS LINE)
import pandas as pd
import altair as alt

df_corg = pd.read_csv('data/corgis_per_country.csv')
df_corg['Years'] = pd.to_datetime(df_corg['Years'].astype('int'), format='%Y')
```

## Melt operation practice

<p>Using the <code>df_corg</code> dataframe, create a new dataframe called <code>df_corg_melt</code> in which you have 'melted' each country column of <code>df_corg</code> over the 'Years' column, into (1) a column in <code>df_corg_melt</code> called 'Country' and (2) where the value for each country column is in a new column in <code>df_corg_melt</code> called 'Births'.</p><p>In other words, your new 'melted' dataframe should have columns called 'Years', 'Country' and 'Births' which lists all combinations of years, countries and number of births.<p><p>You will want to reference what we went over in class/in the extra Altair recordings for these operations and/or the <a href='https://altair-viz.github.io/user_guide/data.html#converting-with-pandas'>documentation about 'melt' in Altair</a> and an example of a <a href='https://altair-viz.github.io/gallery/multiline_tooltip.html#multi-line-tooltip'>multi-line plot using the melt operation with Altair</a>.  (Try with/without using <code>.reset_index()</code>.)</p><p></p><p>You will likely benefit from also reviewing some of the pre-recorded lecture materials.  See the Modules 8.1 and 8.2.(1-3) on the class <a href='https://mediaspace.illinois.edu/channel/IS445%2BOAU_OAG%2BSpring%2B2025/366181802'>MediaSpace page</a>.</p>

```python
#grade (write your code in this cell and DO NOT DELETE THIS LINE)
df_corg_melt = df_corg.melt(id_vars=['Years'], var_name='Country', value_name='Births')
```

## Multi-line Chart

<p>Using <code>df_corg_melt</code>, make a multiline plot of corgis born over time over all 'Years', where each individual line corresponds to a different country.</p><p>Call this chart <code>chart1</code>.<p><p>Keep the default color scheme, but scale the y-axis by a symmetric log.  Use an ordinal scaling for the 'Country' line colors.</p>

```python
#grade (write your code in this cell and DO NOT DELETE THIS LINE)
chart1 = alt.Chart(df_corg_melt).mark_line().encode(
    x=alt.X('Years:T'),
    y=alt.Y('Births:Q', scale=alt.Scale(type='symlog')),
    color=alt.Color('Country:O')
).properties(
    title='Corgis Born Over Time by Country'
)
```

## Multi-line Chart, with Interactivity

<p>We will now update the colormap for the above multiline chart to be something more sensible and include interactivity.</p><p>Using <code>df_corg_melt</code>, make an interactive multiline plot of corgis born over time over all 'Years'.</p><p>Call this chart <code>chart2</code>.<p><p>Use a similar format to the previous line chart except:</p><ul><li>Use the 'sinebow' color scheme for 'Country'.</li><li>Add a dropdown which selects each country's line from a list.  Give the dropdown the <code>name</code> of 'Country: '<ul><li>When the line is selected, the line thickness should be 5, and unselected lines should have a thickness of 1.  </li><li>The default selected 'Country' should be 'United States' </li></ul></li></ul><p>Hint: you might want to check out some of the examples under <a href='https://altair-viz.github.io/gallery/index.html#interactive-charts'>Altair's Interactive Charts</a> for some ideas of how to approach this plot.</p>

```python
#grade (write your code in this cell and DO NOT DELETE THIS LINE)
selection = alt.selection_point(
    fields=['Country'], 
    bind=alt.binding_select(options=df_corg_melt['Country'].unique(), name='Country: '),
    value='United States'
)

chart2 = alt.Chart(df_corg_melt).mark_line().encode(
    x=alt.X('Years:T'),
    y=alt.Y('Births:Q', scale=alt.Scale(type='symlog')),
    color=alt.Color('Country:O', scale=alt.Scale(scheme='sinebow')),
    size=alt.condition(selection, alt.value(5), alt.value(1))
).add_params(
    selection
).properties(
    title='Interactive Corgis Born Over Time by Country'
)

chart1, chart2
```
