```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact, interactive
from IPython.display import display
from matplotlib.colors import Normalize
```

## Plot \#1 Code

```python
url = "https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/ufo-subset-spring2023.csv"
ufo_data = pd.read_csv(url, parse_dates=["date"])

# Check the first few rows to understand the structure
ufo_data.head()


```

```python
# Get column names for categorical and numerical columns
categorical_columns = ufo_data.select_dtypes(include=["object"]).columns.tolist()
numerical_columns = ufo_data.select_dtypes(include=[np.number]).columns.tolist()
print(categorical_columns, "\n", numerical_columns)
```

```python
def create_scatter_plot(x_col, y_col, marker_type, color_col, cmap):
    fig, ax = plt.subplots(figsize=(8, 6))

    # Scatter plot with customizable marker type and color scale
    scatter = ax.scatter(
        ufo_data[x_col], 
        ufo_data[y_col], 
        c=ufo_data[color_col] if color_col else None, 
        cmap=cmap, 
        marker=marker_type
    )
    
    # Add color bar if color_col is selected
    if color_col:
        plt.colorbar(scatter)
    
    ax.set_xlabel(x_col)
    ax.set_ylabel(y_col)
    ax.set_title(f'Scatter Plot: {x_col} vs {y_col}')
    
    plt.show()
```

```python
# Dropdown widgets for user selection in Scatter Plot
x_dropdown = widgets.Dropdown(options=numerical_columns, description='X Axis:')
y_dropdown = widgets.Dropdown(options=numerical_columns, description='Y Axis:')
marker_dropdown = widgets.Dropdown(options=['o', 's', 'D', '^'], description='Marker Type:')
color_dropdown = widgets.Dropdown(options=[None] + list(numerical_columns), description='Color by:')
cmap_dropdown = widgets.Dropdown(options=['viridis', 'jet', 'plasma'], description='Colormap:')

# Use interact to link the function with the widgets
interact(create_scatter_plot, 
         x_col=x_dropdown, 
         y_col=y_dropdown, 
         marker_type=marker_dropdown, 
         color_col=color_dropdown, 
         cmap=cmap_dropdown);
```

## Plot \#1 Write Up

For Plot 1, I created a scatter plot where users can select different numerical columns for the x and y axes. The plot allows the user to change the marker type and the color of the markers based on another column in the dataset. The color scale of the scatter plot is adjustable through a dropdown that lets the user pick from predefined colormap options such as 'viridis,' 'jet,' and 'plasma'. The user can also choose a categorical column to represent the color scale or leave it as default without coloring the markers.

The scatter plot allow flexibility in analyzing different variables in the UFOs dataset. By letting users pick from various combinations of numerical columns and markers, it offer insight into the relationships between different variables, such as the correlation between reported UFO sightings' duration and location. Also, the user can adjust the colormap to better represent the dataset's range of values...

## Plot \#2 Code

```python
def create_binned_plot(bin_col, num_bins):
    fig, ax = plt.subplots(figsize=(10, 6))  # Adjusted figure size

    if bin_col in numerical_columns:
        # Histogram for numerical columns
        ax.hist(ufo_data[bin_col], bins=num_bins, edgecolor='black', color='skyblue')
        ax.set_xlabel(bin_col)
        ax.set_ylabel('Frequency')
        ax.set_title(f'Histogram of {bin_col} with {num_bins} bins')
    else:
        # Bar chart for categorical columns (display top 10 categories)
        value_counts = ufo_data[bin_col].value_counts().head(10)  # Top 10 categories
        value_counts.plot(kind='bar', ax=ax, color='skyblue')
        ax.set_xlabel(bin_col)
        ax.set_ylabel('Count')
        ax.set_title(f'Top 10 Categories of {bin_col}')
        
        # Rotate x-axis labels for better readability
        ax.set_xticklabels(value_counts.index, rotation=45, ha='right')

    # Add gridlines for better visualization
    ax.grid(axis='y', linestyle='--', alpha=0.7)

    plt.tight_layout()  # Adjust layout to prevent clipping
    plt.show()

# Dropdown widgets for user selection in Histogram Plot
bin_column_dropdown = widgets.Dropdown(options=categorical_columns + numerical_columns, description='Bin Column:')
num_bins_slider = widgets.IntSlider(min=1, max=50, step=1, value=10, description='Number of Bins:')

# Use interact to link the function with the widgets
interact(create_binned_plot, 
         bin_col=bin_column_dropdown, 
         num_bins=num_bins_slider)
```
