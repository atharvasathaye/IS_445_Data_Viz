## Plot \#1 Code

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
from ipywidgets import interact

url = "https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/licenses_fall2022.csv"
df = pd.read_csv(url)

numerical_cols = df.select_dtypes(include=['number']).columns.tolist()
categorical_cols = df.select_dtypes(exclude=['number']).columns.tolist()

x_dropdown = widgets.Dropdown(options=numerical_cols + categorical_cols, description="X-axis")
y_dropdown = widgets.Dropdown(options=numerical_cols + categorical_cols, description="Y-axis")
marker_dropdown = widgets.Dropdown(options=["o", "s", "D", "^"], description="Marker")
color_dropdown = widgets.Dropdown(options=numerical_cols + categorical_cols, description="Color")

def update_plot(x, y, marker, color):
    plt.figure(figsize=(8, 6))

    df_plot = df.copy()  # Create a copy of the DataFrame

    if x in categorical_cols:
        df_plot[x] = df_plot[x].astype('category').cat.codes
    if y in categorical_cols:
        df_plot[y] = df_plot[y].astype('category').cat.codes
    
    if color in categorical_cols:
        df_plot[color] = df_plot[color].astype('category').cat.codes
        cmap = 'Set1'  # Use a discrete colormap for categories
        colorbar_label = f"{color} (Categorical Codes)"
    else:
        cmap = 'viridis'
        colorbar_label = color

    scatter = plt.scatter(df_plot[x], df_plot[y], c=df_plot[color], cmap=cmap, marker=marker, alpha=0.7)
    plt.colorbar(scatter, label=colorbar_label)
    plt.xlabel(x)
    plt.ylabel(y)
    plt.title("Interactive Scatter Plot")
    plt.show()

interact(update_plot, x=x_dropdown, y=y_dropdown, marker=marker_dropdown, color=color_dropdown);
```

## Plot \#1 Write Up

This scatter plot was created using matplotlib, seaborn, and ipywidgets to allow interactive selection of x and y axes, marker type, and color mapping. The user can choose both categorical and numerical columns for x and y axes, with visualizations adapting accordingly. Categorical variables are visualized using a strip plot, while numerical variables use a scatter plot with a color gradient.One challenge was ensuring compatibility when mixing categorical and numerical columns, particularly with color mapping.

## Plot \#2 Code

```python
bin_dropdown = widgets.Dropdown(options=numerical_cols + categorical_cols, description="Bin Column")
bin_slider = widgets.IntSlider(min=1, max=50, step=1, value=10, description="Bins")

MAX_BINS = 30

def update_histogram(column, bins):
    plt.figure(figsize=(9, 6))
    if column in numerical_cols and column not in ['LICENSE NUMBER']:
        num_bins = min(bins, MAX_BINS)
        sns.histplot(df[column].dropna(), bins=num_bins, kde=False, color='royalblue', edgecolor='black')
        plt.xlabel(column, fontsize=12)
        plt.ylabel("Count", fontsize=12)
        plt.title(f"Histogram of {column}", fontsize=14)
    elif column in categorical_cols:
        value_counts = df[column].value_counts()
        top_categories = value_counts.index[:20]
        filtered_df = df[df[column].isin(top_categories)]
        categorical_order = pd.Categorical(filtered_df[column], categories=top_categories, ordered=True)
        sns.countplot(y=categorical_order, hue=categorical_order, palette='coolwarm', legend=False)
        plt.xlabel("Count", fontsize=12)
        plt.ylabel(column, fontsize=12)
        plt.title(f"Bar Chart of {column}", fontsize=14)
    else:
        plt.text(0.5, 0.5, "Histogram not available for this column", fontsize=12, ha='center')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

interact(update_histogram, column=bin_dropdown, bins=bin_slider);
```

## Plot \#2 Write Up

This plot was constructed using matplotlib and ipywidgets to create an interactive histogram and bar chart. The user can select a numerical or categorical column to visualize. If a numerical column is chosen, a histogram is generated with a user-defined number of bins.If a categorical column is selected, a bar chart is displayed with counts for each category.
