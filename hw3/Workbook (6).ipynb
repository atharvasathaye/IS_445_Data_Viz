## Question \#1

Create a vertical integer widget slider that starts at 100 and ends at 200.  Use the `description` key to name this widget "Int Slider".  **Call this slider "islider".**

See the [IntSlider Docs](https://ipywidgets.readthedocs.io/en/stable/examples/Widget%20List.html#intslider) for more information.

```python
#grade (write your code in this cell and DO NOT DELETE THIS LINE)
import ipywidgets as widgets
islider=widgets.IntSlider(
    value=100,
    min=100,
    max=200,
    step=1,
    description = "Int Slider",
    orientation='vertical',
)
islider
```

## Question \#2

Create a [ToggleButton](https://ipywidgets.readthedocs.io/en/stable/examples/Widget%20List.html#togglebutton) that has a "success" style, and a description of "Click me!".  Use the "Hippo" [font-awesome](https://fontawesome.com/icons/hippo?f=classic&s=solid) (use `icon='hippo'`).

**Call this button "togbutton".**

```python
#grade (write your code in this cell and DO NOT DELETE THIS LINE)
togbutton=widgets.ToggleButton(
    value=False,
    description="Click me!",
    button_style='success',
    icon='hippo'
    
)
togbutton
```

## Question \# 3

Create a [Dropdown](https://ipywidgets.readthedocs.io/en/stable/examples/Widget%20List.html#dropdown) with options of ['hi', 3, 'stuff'] and an initial value of 3.  The description should be "Options:".

**Call this button "dropdown".**

```python
#grade (write your code in this cell and DO NOT DELETE THIS LINE)
dropdown=widgets.Dropdown(
    options=['hi',3,'stuff'],
    value=3,
    description='Options:',
    disabled=False
)
dropdown
```
