```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
from ipywidgets import interact

url = "https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/licenses_fall2022.csv"
df = pd.read_csv(url)

numerical_cols = df.select_dtypes(include=['number']).columns.tolist()
categorical_cols = df.select_dtypes(exclude=['number']).columns.tolist()

x_dropdown = widgets.Dropdown(options=numerical_cols + categorical_cols, description="X-axis")
y_dropdown = widgets.Dropdown(options=numerical_cols + categorical_cols, description="Y-axis")

marker_dropdown = widgets.Dropdown(options=["o", "s", "D", "^"], description="Marker")

color_dropdown = widgets.Dropdown(options=numerical_cols + categorical_cols, description="Color")

def update_plot(x, y, marker, color):
    plt.figure(figsize=(8, 6))
    
    if x in categorical_cols:
        df[x] = df[x].astype('category').cat.codes
    if y in categorical_cols:
        df[y] = df[y].astype('category').cat.codes
    if color in categorical_cols:
        df[color] = df[color].astype('category').cat.codes
    
    scatter = plt.scatter(df[x], df[y], c=df[color], cmap='viridis', marker=marker, alpha=0.7)
    plt.colorbar(scatter, label=color)
    plt.xlabel(x)
    plt.ylabel(y)
    plt.title("Interactive Scatter Plot")
    plt.show()

interact(update_plot, x=x_dropdown, y=y_dropdown, marker=marker_dropdown, color=color_dropdown);
```

## Plot \#2 Code

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
from ipywidgets import interact

url = "https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/licenses_fall2022.csv"
df = pd.read_csv(url)

numerical_cols = df.select_dtypes(include=['number']).columns.tolist()
categorical_cols = df.select_dtypes(exclude=['number']).columns.tolist()

bin_dropdown = widgets.Dropdown(options=numerical_cols + categorical_cols, description="Bin Column")
bin_slider = widgets.IntSlider(min=1, max=50, step=1, value=10, description="Bins")

def update_histogram(column, bins):
    plt.figure(figsize=(8, 6))
    if column in numerical_cols:
        plt.hist(df[column].dropna(), bins=bins, alpha=0.7, color='blue', edgecolor='black')
        plt.xlabel(column)
        plt.ylabel("Count")
        plt.title("Histogram of " + column)
    else:
        df[column].value_counts().plot(kind='bar', color='blue', edgecolor='black')
        plt.xlabel(column)
        plt.ylabel("Count")
        plt.title("Bar Chart of " + column)
    plt.show()

interact(update_histogram, column=bin_dropdown, bins=bin_slider);
```
