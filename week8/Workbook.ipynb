```
import pandas as pd
import altair as alt
import os
import warnings

warnings.filterwarnings("ignore", category=FutureWarning)
alt.data_transformers.disable_max_rows()
os.makedirs("assets", exist_ok=True)
```

```
url = 'https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/bfro_reports_fall2022.csv'
df = pd.read_csv(url)
```

```
df['year'] = pd.to_datetime(df['date'], errors='coerce').dt.year
df.dropna(subset=['latitude', 'longitude'], inplace=True)

sightings_per_year = df.groupby('year').size().reset_index(name='count')

bar_chart = alt.Chart(sightings_per_year).mark_bar().encode(
    x=alt.X('year:O', title='Year'),
    y=alt.Y('count:Q', title='Number of Sightings'),
    tooltip=['year', 'count']
).properties(
    width=600,
    height=400,
    title="Number of Bigfoot Sightings by Year"
)

bar_chart.save('assets/plot1.json')
bar_chart
```

```
df_scatter = df.dropna(subset=['season'])

scatter = alt.Chart(df_scatter).mark_circle(size=30).encode(
    x=alt.X('longitude:Q', title='Longitude'),
    y=alt.Y('latitude:Q', title='Latitude'),
    color=alt.Color('season:N', title='Season'),
    tooltip=['state', 'location_details', 'date', 'season']
).properties(
    width=700,
    height=450,
    title='Bigfoot Sightings by Location and Season'
)

scatter.save('assets/plot2.html')
scatter.save('assets/plot2.json')
scatter
```
