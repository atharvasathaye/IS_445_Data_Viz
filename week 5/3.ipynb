## Step 1: Look at toy dataset

We are going to use the data frame, `df`, below to make a [pivot table](https://pandas.pydata.org/docs/reference/api/pandas.pivot_table.html).

First, take a look at this `df` dataframe to get familiar with it:

```python
#grade (keep this code in this cell and DO NOT DELETE THIS LINE) -- this code should show up in your Lab!
import pandas as pd

df = pd.DataFrame({
    'A':['bar', 'bar', 'bar', 'bar', 'bar', 'bar', 'foo', 'foo', 'foo', 'foo', 'foo', 'foo', 'foobar', 'foobar', 'foobar', 'foobar', 'foobar', 'foobar', 'foobar', 'foobar'],
    'B':['one', 'one', 'two', 'three', 'two', 'three', 'one', 'three', 'one', 'two', 'one', 'one', 'three', 'three', 'two', 'three', 'one', 'one', 'three', 'two'],
    'C':['small', 'medium', 'medium', 'medium', 'medium', 'medium', 'small', 'small', 'medium', 'small', 'large', 'small', 'small', 'large', 'medium', 'large', 'medium', 'medium', 'large', 'large'],
    'D':[15.0, 7.0, 1.0, 6.0, 102.0, 88.0, 88.0, 6.0, 102.0, 6.0, 12.0, 102.0, 6.0, 102.0, 15.0, 1.0, 88.0, 6.0, 88.0, 14.0],
    'E':[11.9, 22.0, 8.0, 9.0, 8.0, 8.0, 2.0, 8.0, 8.0, 5.0, 9.0, 8.0, 8.0, 9.0, 9.0, 9.0, 2.0, 9.0, 9.0, 9.0]
})


df
```

## Step 2: Create pivot table

Create a pivot table such that the dataframe will have the mean of all of the values of column "E" into bins of columns "B" and "A" where "B" values will be the row name and "A" values will be the column name.

Set the parameter `fill_value = np.nan` to fill in NaN's for missing values in the table.

```python
#grade (write your code in this cell and DO NOT DELETE THIS LINE)
import numpy as np
table = pd.pivot_table(df, values='E', index=['B'], columns=['A'], aggfunc='mean', fill_value=np.nan)

print(table)
```

## Step 3: Create Heatmap

Create a gridheatmap figure with the "Greens" colorscheme using `bqplot`.

Label the x-axis with 'A' and the yaxis with 'B' and place the color axis on the right side of the plot.

The `color` of this grid heatmap should be the *values* of the table (`table.values`) and the rows should be the names of the rows of your table, stored in a variable **called `row_names`**, and the columns should be the columns of your table stored in a variable **called `column_names`**.
 * Hint \#1: Check the pivot table Lab section for a reminder of how to get these from your table
 * Hint \#2: Not sure how to set row and column values?  Try looking at the doc strings with `bqplot.GridHeatMap?`

Add in an interaction in which you `select` when you `click`.  When a cell is selected, fill the cell with the color magenta.

Place this heatmap mark in a figure.

**Call the heatmap variable `heatmap` and the figure object `fig`.**

```python
#grade (write your code in this cell and DO NOT DELETE THIS LINE)
import bqplot as bqplot
import ipywidgets as widgets

row_names = table.index.astype('str').values
column_names = table.columns.astype('str').values

col_sc = bqplot.ColorScale(scheme='Greens')
x_sc = bqplot.OrdinalScale()
y_sc = bqplot.OrdinalScale()

col_ax = bqplot.ColorAxis(scale=col_sc, side='right')
x_ax = bqplot.Axis(label='A', scale=y_sc, side='bottom')
y_ax = bqplot.Axis(label='B', orientation='vertical', scale=x_sc)

heatmap = bqplot.GridHeatMap(color=table.values,
                            row=row_names,
                            column=column_names,
                            scales={'color': col_sc, 'row': y_sc, 'column': x_sc},
                            interactions={'click': 'select'},
                            selected_style={'fill': 'magenta'})

fig = bqplot.Figure(marks=[heatmap], axes=[x_ax, y_ax, col_ax])


print(heatmap,fig)
```

## Step 4: Create a histogram of column "E" with a "hard-coded" cell selection

### Step 4.1: Get row/column names with hardcoded values

Set an integer **called `i`** to the cell value 0 and another integer **called `j`** and set it to 2.

Get the column entry associated with this `i` value and set it to a **variable called `column_select_name`** and get the row entry associated with this `j` value and set it to a **variable called `row_select_name`**.

```python
#grade (write your code in this cell and DO NOT DELETE THIS LINE)
i = 0
j = 2
column_select_name = column_names[i]
row_select_name = row_names[j]

print(column_select_name)
print(row_select_name)
```

### Step 4.2: Hardcoded subset dataframe

Create a subset dataframe **called `df_subset`** which is a subset of `df` that contains only the data where the "A" column is equal to `column_select_name` and "B" column is equal to `row_select_name`.

Hint: not sure how to subset?
  * check out the [`.isin` function](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.isin.html)
  * there are two conditions that need to be met -- you can do this by making `df_subset` in two steps or using a [boolean `&` symbol](https://stackoverflow.com/questions/21415661/logical-operators-for-boolean-indexing-in-pandas)

```python
#grade (write your code in this cell and DO NOT DELETE THIS LINE)
df_subset = df[(df['A'] == column_select_name) & (df['B'] == row_select_name)]

print(df_subset)
```

### Step 4.3: Put it all together in a function

Create a function **called `subsetting_function`** that takes as inputs **in this order** (1) a dataframe, (2) the column names of your grid heatmap, (3) the row names of your grid heatmap, (4) a cell number for the columns (like `i`) and (5) a cell number for the rows (like `j`) and outputs the subset dataframe that results from this selection.

```python
#grade (write your code in this cell and DO NOT DELETE THIS LINE)
def subsetting_function(dataframe, column_names, row_names, i, j):
    column_select_name = column_names[i]
    row_select_name = row_names[j]
    df_subset = dataframe[(dataframe['A'] == column_select_name) & (dataframe['B'] == row_select_name)]
    return df_subset
```

### Step 4.4: Use function to get a data subset for plotting

Using your `subsetting_function`, create a subset dataframe **called `df_subset_plot`** for row, column = 0,1.

```python
#grade (write your code in this cell and DO NOT DELETE THIS LINE)
df_subset_plot = subsetting_function(df, column_names, row_names, 1, 0)

print(df_subset_plot)
```

### Step 4.5: Histogram with this subset dataframe

Using the bqplot `Bars` mark, create a plot of the *mean* of column "D" for each unique value of column "C" in `df_subset_plot`.  **Call this bar mark `bars`.**  Put this bar mark into a figure **called `fig_bars`** and label the x-axis as "C" and the y-axis as "Mean D".

Hints:
 * not sure of the parameters of `Bars`?  try `bqplot.Bars?`
 * not sure about how to "deposit" 'D' onto 'C'?  What function can you us in Pandas for this? (Check out what we did for the Pandas lab).  What are the `.index` and `.values` of this series?
 * not sure what to use for scales? What type of variable is `df`'s "C" column (x-scale)? What type is the "D" column (y-scale)?

```python
#grade (write your code in this cell and DO NOT DELETE THIS LINE)
mean_d = df_subset_plot.groupby('C')['D'].mean()

x_sc_bars = bqplot.OrdinalScale()
y_sc_bars = bqplot.LinearScale()

x_ax_bars = bqplot.Axis(scale=x_sc_bars, label='C', side='bottom')
y_ax_bars = bqplot.Axis(scale=y_sc_bars, orientation='vertical', label='Mean D', side='left')

bars = bqplot.Bars(x=mean_d.index.values, y=mean_d.values, scales={'x': x_sc_bars, 'y': y_sc_bars})

fig_bars = bqplot.Figure(marks=[bars], axes=[x_ax_bars, y_ax_bars])
```

## Step 5: Create selection function

Create a function **called `on_selection`** which is tied to the heatmap with an `.observe` call on the `heatmap`'s `selected` key.

The function should only allow for *one* selection at a time (i.e., no multiple cell selection).

The function should change the data used in the `bars` plot based on the cell selected by the user.

**NOTE:** Be sure to account for the fact that some cell's will result in empty subset dataframes -- this will lead to an error.  To fix this, make sure the `bars` plot has the value of an empty string (`['']`) on its x-axis ("C") and the value of `[0]` on the y-axis ("Mean C") when the subset dataframe is an empty string.

```python
#grade (write your code in this cell and DO NOT DELETE THIS LINE)
def on_selection(change):
    if heatmap.selected is not None and len(heatmap.selected) == 1:
        col, row = heatmap.selected[0]
        df_subset_plot = subsetting_function(df, column_names, row_names, col, row)
        if not df_subset_plot.empty:
            mean_d_subset = df_subset_plot.groupby('C')['D'].mean()
            bars.x = mean_d_subset.index.values
            bars.y = mean_d_subset.values
        else:
            bars.x = ['']
            bars.y = [0]
    else:
        df_subset_plot = subsetting_function(df, column_names, row_names, 1, 0)
        mean_d = df_subset_plot.groupby('C')['D'].mean()
        bars.x = mean_d.index.values
        bars.y = mean_d.values

heatmap.observe(on_selection, 'selected')
```

## Step 6: Put it all together in a dashboard

Create a dashboard variable **called `dashboard`**.  On the left side of this dashboard place the `fig` heatmap figure and on the right place the bar plot figure, `fig_bars`.

```python
#grade (write your code in this cell and DO NOT DELETE THIS LINE)
dashboard = widgets.HBox([fig, fig_bars])

print(dashboard)
```
