## Step 1: Look at toy dataset

We are going to use the data frame, `df`, below to make a [pivot table](https://pandas.pydata.org/docs/reference/api/pandas.pivot_table.html).

First, take a look at this `df` dataframe to get familiar with it:

```python
#grade (keep this code in this cell and DO NOT DELETE THIS LINE) -- this code should show up in your Lab!
import pandas as pd

df = pd.DataFrame({
    'A':['bang', 'bar', 'bar', 'bar', 'bar', 'foo', 'foo', 'foo'],
    'B':['one', 'three', 'three', 'one', 'three', 'one', 'one', 'one'],
    'C':['small', 'small', 'large', 'medium', 'medium', 'medium', 'small', 'medium'],
    'D':[7, 6, 2, 7, 2, 7, 2, 1],
    'E':[11, 2, 2, 2, 2, 11, 2, 11]
})


df
```

## Step 2: Pivot Table Walkthrough

Next, let's make a pivot table with this dataset such that the resulting dataframe will have the sum of all of the values of column "E", into bins of columns "B" and "A" where unique "B" values will the the row names and unique "A" values will be the column names.

First, check out some of the options with Pandas pivot tables below by running the following cell to show the doc strings:

```python
pd.pivot_table?
```

The important parts to focus on in the above are the "values", "index" and "columns" parameters -- we just need to figure out which correspond to columns "A", "C" and "D" in the example printed out in the doc string above.

Additionally, the "aggfunc" parameter tells us how we are going to aggregate the data.

Create a variable **called `table1`** using `df` and the `pd.pivot_table` function where you set the following parameters:
 * `values = 'E'`
 * `index = ['B']`
 * `columns = ['A']`
 * `aggfunc = numpy.sum` (don't forget to load the numpy library!)

```python
#grade (write your code in this cell and DO NOT DELETE THIS LINE)
import numpy as np
table1 = pd.pivot_table(df, values='E', index=['B'], columns=['A'], aggfunc='sum')
print(table1)
```

## Step 3: Second Pivot table

Now, make a pivot table **called `table2`** such that the resulting dataframe will have the mean of all values of column "D" into bins of columns "A" and "C" where unique "A" values will the the row names and unique "C" values will be the column names.

```python
#grade (write your code in this cell and DO NOT DELETE THIS LINE)
table2 = pd.pivot_table(df, values='D', index=['A'], columns=['C'], aggfunc='mean')
print(table2)
```

## Step 3: Getting Row and Column labels

The final thing we likely want to do is to be able to get the row and column labels from our pivot table for future use in something like a heatmap as row/column labels.

Create a variable **called `row_names`** and set it equal to `table2.index.values.astype('str')`.  

Additionally, create a variable **called `column_names`** and set it equal to `table2.columns.values.astype('str')`.  

```python
#grade (write your code in this cell and DO NOT DELETE THIS LINE)
row_names = table2.index.astype('str').values
column_names = table2.columns.astype('str').values

print(row_names)
print(column_names)
```
