## Step 2: Create color, x, y scales

Create a color scale with the "Blues" color scheme.  **Call this variable `colsc`.**

Create an x-scale and **call this variable `xsc`**.  Hint: what [type of scale](https://bqplot.github.io/bqplot/api/scales/) should you use for this variable?

Create an y-scale and **call this variable `ysc`**.  Hint: same question of scale type for the x-scale.


````python
#grade (write your code in this cell and DO NOT DELETE THIS LINE)
import bqplot
from bqplot import pyplot as plt

colsc = bqplot.ColorScale(scheme='Blues')
xsc = bqplot.OrdinalScale()
ysc = bqplot.OrdinalScale()

print(colsc)
print(xsc)
print(ysc)
````

## Step 3: Create color, x, y axis

Create a color axis **called this variable `colax`.**  Place this color axis on the top of the plot, horizontally.

Create an x-axis and **call this variable `xax`**.  Label this axis with the text "X-Variable".

Create an y-axis and **call this variable `yax`**.  Label this axis with the text "Y-Variable".

Place the x/y axis in the "usual" places for these types of axis.

**NOTE:** for testing purposes, make sure all of the placements/choices are explicit (i.e., don't rely on the defaults for any of these objects).

````python
#grade (write your code in this cell and DO NOT DELETE THIS LINE)
colax = bqplot.ColorAxis(scale=colsc, orientation='horizontal', side='top')
xax = bqplot.Axis(scale=xsc, orientation='horizontal', label='X-Variable', side='bottom')
yax = bqplot.Axis(scale=ysc, orientation='vertical', label='Y-Variable', side='left')

print(colax)
print(xax)
print(yax)
````

## Step 4: Create heat map

Create a "GridHeatMap" and **call this variable `heatmap`** using your previously created `colsc`, `xsc` and `ysc` scales.

The `color` of this grid heatmap should be the *mean* of `data3d` on its 3rd axis (i.e., along the "longest" axis).

Add in an interaction in which you `select` when you `click`.  When a cell is selected, fill the cell with the color red.

````python
#grade (write your code in this cell and DO NOT DELETE THIS LINE)
heatmap = bqplot.GridHeatMap(color=np.mean(data3d, axis=2),
                                scales={'color': colsc, 'row': ysc, 'column': xsc},
                                interactions={'click': 'select'},
                                selected_style={'fill': 'red'})

print(heatmap)
````

## Step 7: Create the heatmap figure

Combine the heatmap and appropriate axes into a `bqplot` figure object.  **Call this variable `fig`.**

````python
#grade (write your code in this cell and DO NOT DELETE THIS LINE)
fig = bqplot.Figure(marks=[heatmap], axes=[colax, xax, yax])

print(fig)
````
